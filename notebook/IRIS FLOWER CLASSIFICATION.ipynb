### 1) Problem statement
- This project aims to classify Iris flower species based on morphological features such as sepal and petal measurements.
- The objective is to build a deep learning model that can accurately predict the species:
    Setosa
    Versicolor
    Virginica
- The system also focuses on providing interpretability using feature importance and evaluation metrics.


### 2) Data Collection
- Dataset Source: Iris Dataset (UCI Machine Learning Repository)
- The dataset contains 150 samples with 4 input features and 1 target variable
